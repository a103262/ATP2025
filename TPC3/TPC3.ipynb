```python
### TPC3: Aplicação para manipulação de listas de inteiros
import random   # Biblioteca para gerar números aleatórios

# Lista global onde vão ser guardados os números
lista = []

# Opção 1 - Criar lista com número aleatório
def criarLista():
    num = random.randint(1, 100)   # escolhe número entre 1 e 100
    lista.append(num)              # adiciona à lista global
    print("Lista criada:", lista)

# Opção 2 - Ler lista introduzida pelo utilizador
def lerLista():
    entrada = input("Introduza os números separados por espaço: ")
    for x in entrada.split():      # percorre cada valor escrito
        lista.append(int(x))       # transforma em inteiro e adiciona
    print("Lista lida:", lista)

# Opção 3 - Soma
def soma():
    if len(lista) == 0:            # se a lista está vazia
        print("A lista está vazia!")
        return
    total = 0
    for num in lista:              # percorre a lista e acumula
        total = total + num
    print("Soma =", total)

# Opção 4 - Média
def mediaLista():
    if len(lista) == 0:
        print("A lista está vazia!") 
        return
    total = 0
    for num in lista:
        total = total + num
    media = total / len(lista)     # divide pelo tamanho da lista
    print("Média =", media)

# Opção 5 - Maior valor
def maiorLista():
    if len(lista) == 0:
        print("A lista está vazia!")
        return
    maior = lista[0]               # começa com o primeiro elemento
    for elem in lista:
        if elem > maior:           # se encontrar maior, atualiza
            maior = elem
    print("Maior =", maior)

# Opção 6 - Menor valor
def menorLista():
    if len(lista) == 0:
        print("A lista está vazia!")
        return
    menor = lista[0]               # começa com o primeiro elemento
    for elem in lista:
        if elem < menor:           # se encontrar menor, atualiza
            menor = elem
    print("Menor =", menor)


# (7) Verifica se está ordenada por ordem crescente
def estaOrdenadaCrescente():
    if len(lista) == 0:
        print("A lista está vazia!")
        return
    for i in range(len(lista)-1):
        if lista[i] > lista[i+1]:  # se encontrar um fora de ordem
            print("Não")           # não está ordenada
            return
    print("Sim")                   # se passou pelo ciclo, está ordenada

# (8) Verifica se está ordenada por ordem decrescente
def estaOrdenadaDecrescente():
    if len(lista) == 0:
        print("A lista está vazia!")
        return
    for i in range(len(lista)-1):
        if lista[i] < lista[i+1]:
            print("Não")
            return
    print("Sim")

# (9) Procura um elemento e mostra a posição
def procuraElemento():
    if len(lista) == 0:
        print("A lista está vazia!")
        return
    elem = int(input("Elemento a procurar: "))
    for i in range(len(lista)):
        if lista[i] == elem:
            print("Elemento encontrado na posição", i)
            return
    print("-1 (não encontrado)")

def menu():
    continuar = "s"  # variável de controlo para continuar ou sair
    while continuar.lower() == "s":
        print("\n### Menu ###")
        print("(1) Criar Lista")
        print("(2) Ler Lista")
        print("(3) Soma")
        print("(4) Média")
        print("(5) Maior")
        print("(6) Menor")
        print("(7) Está ordenada por ordem crescente?")
        print("(8) Está ordenada por ordem decrescente?")
        print("(9) Procurar um elemento")
        print("(0) Sair")
        
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            criarLista()
        elif opcao == "2":
            lerLista()
        elif opcao == "3":
            soma()
        elif opcao == "4":
            mediaLista()
        elif opcao == "5":
            maiorLista()
        elif opcao == "6":
            menorLista()
        elif opcao == "7":
            estaOrdenadaCrescente()
        elif opcao == "8":
            estaOrdenadaDecrescente()
        elif opcao == "9":
            procuraElemento()
        elif opcao == "0":
            continuar = "n"  # muda a variável para sair
            print("Aplicação terminada. Lista final:", lista)
        else:
            print("Opção inválida, tente novamente!")

        # Pergunta se quer continuar, apenas se não escolheu sair
        if opcao != "0":
            continuar = input("Quer continuar? (s/n): ")

# Programa principal
menu()
```
