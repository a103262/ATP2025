### TPC5: Aplicação para gestão de alunos

Considere que o modelo do aluno e da turma têm a seguinte estrutura:

`aluno = (nome, id, [notaTPC, notaProj, notaTeste])`

`turma = [aluno]`

* Cria uma aplicação que coloca no monitor o seguinte menu de operações:
    - 1: Criar uma turma;
    - 2: Inserir um aluno na turma;
    - 3: Listar a turma;
    - 4: Consultar um aluno por id;
    - 8: Guardar a turma em ficheiro;
    - 9: Carregar uma turma dum ficheiro;
    - 0: Sair da aplicação
* No fim de executar a operação selecionada, a aplicação deverá colocar novamente o menu e pedir ao utilizador a opção para continuar;
* Utiliza a tua aplicação para criar uma turma com 5 alunos.

```python
#aluno = (nome, id, [notaTPC, notaProj, notaTeste])
#turma = [aluno]

# criar uma turma 
def criarTurma():
    turma = []
    print("Turma criada.")
    return turma

#inserir aluno da turma 
def inserirAluno(turma):
    for aluno in turma:
        nome= input("introduza o nome de aluno: ")
        id = int(input("introduza o numero de aluno: "))
        notaTPC = float(input("introduza a nota do tpc: "))
        notaProj = float(input("introduza a nota do tpc: "))
        notaTeste = float(input("introduza a nota do tpc: "))
        lista = [notaTPC, notaProj, notaTeste]
        aluno = (nome, id, lista)
        turma.append(aluno)
    return turma 

#listar a turma 
def listarTurma(turma):
    if not turma:
        print("A turma está vazia.\n")
        return
    for alunos in turma:
        nome, id, notas = alunos  # Desempacotar o tuplo
        print(f'Nome: {nome}, ID: {id}, Notas: {notas}')
    return

#consultar aluno por id

def consultarId(turma,id):
    i = 0
    id = int(input('Introduza o id do aluno: '))
    # print(turma.ennumerate(aluno))  # ERRADO: 'enumerate' não é método de listas
# Uso correto: for i, aluno in enumerate(turma):
    for i, aluno in enumerate(turma): # i corresponde ao índice de cada aluno dentro da turma
        if aluno[1] == id:
            print(f'Nome: {aluno[0]}, ID: {id}, Notas: {aluno[2]}')
            return
    print(f'Aluno com id {id} não encontrado na turma')


#guardar turma  em ficheiro 
def guardarTurma(turma, fnome):
    with open(fnome, "w") as file:
        for nome, id, notas in turma:
            # Formata o aluno como "nome,id::nota1,nota2,nota3"
            notas_str = ",".join(map(str, notas))
            linha = f"{nome},{id}::{notas_str}\n"
            file.write(linha)
    print(f'Turma guardada em {fnome}')



#carregar a turma de um ficheiro
def carregarFicheiro(fnome):
    turma = []
    file = open(fnome,"r")
    for linha in file:
        if linha != '': # Identifica linhas com texto
            dados = linha.split('::') # Dividimos as linhas do ficheiro nos dados do tuplo
            nome_id = dados[0].split(",")
            nome = nome_id[0]
            id = int(nome_id[1])
            notas = list(map(int, dados[1].split(",")))
            aluno = (nome, id, notas)
            turma.append(aluno)
    return turma


#sair da aplicacao 
def sairAplicacao():
    print("A sair da aplicação... Até logo!")
    exit()
    
    
# Variável onde vamos armazenar os dados da turma
turma = []

def menu():
    print("""
--- MENU ---
1: Criar uma turma
2: Inserir um aluno na turma
3: Listar a turma
4: Consultar um aluno por id
5: Guardar a turma em ficheiro
6: Carregar uma turma dum ficheiro
0: Sair da aplicação
""")

opcao = -1

while opcao != 0:
    menu()
    texto = input("Selecione a opção desejada: ").strip()
    if texto == "":
        print(" Tem de introduzir um número!\n")
        continue
    try:
        opcao = int(texto)
    except ValueError:
        print(" Introduza um número válido!\n")
        continue

    if opcao == 1:
        turma = criarTurma()
    elif opcao == 2:
        inserirAluno(turma)
    elif opcao == 3:
        listarTurma(turma)
    elif opcao == 4:
        consultarId(turma)
    elif opcao == 5:
        guardarTurma(turma, "turma.txt")
    elif opcao == 6:
        turma = carregarFicheiro("turma.txt")
    elif opcao == 0:
        print("Até à próxima!")
    else:
        print("Opção inválida!\n")
```
