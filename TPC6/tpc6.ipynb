```python
# TabMeteo = [(Data,TempMin,TempMax,Precipitacao)]
    # Data = (Int,Int,Int)
    # TempMin = Float
    # TempMax = Float
    # Precipitacao = Float

def menu():
    print('\n Menu')
    print('1) Média das temperaturas')
    print('2) Guardar tabela meteorológica')
    print('3) Carregar tabela meteorológica')
    print('4) Temperatura mínima mais baixa')
    print('5) Amplitude máxima')
    print('6) Dia com maior precipitação')
    print('7) Dias com precipitação superior a um limite')
    print('8) Maior período de calor')
    print('9) Mostrar gráfico meteorológico')
    print('0) Sair')


def medias(tabMeteo1):
    res = []
    for data, tmin, tmax, prec in tabMeteo1:
        media = (tmin + tmax) /2
        res.append((data,media))    
    return res


def guardaTabMeteo(t, fnome):
    f = open(fnome, "w")
    for data, tmin, tmax, prec in t:
        ano, mes, dia = data
        registo = f"{ano}-{mes}-{dia}|{tmin}|{tmax}|{prec}\n"
        f.write(registo)
    f.close() #nao esquecer disto
    return   


def carregaTabMeteo(fnome):
    res = []
    f = open(fnome, "r")
    #text = f.read()
    #strip tira espacos em branco 
    for linha in f:
        linha = linha.strip()
        campos = linha.split("|")
        #data = campos[0]
        #tmin= campos[1] #fazer isto é igual ao que esta abaixo :
        data, tmin, tmax, prec = campos 
        ano, mes, dia = data.split("-")
        #resultado split é smp strings 
        res.append(((int(ano),int(mes),int(dia)),float(tmin),float(tmax),float(prec)))
        print(campos)
    f.close()
    return res


def minMin(tabMeteo1):
    i = 1
    minima = tabMeteo1[0][1] # começa pelo primeiro valor
    if tabMeteo1[i][1] < minima: # compara com o valor mínimo atual
            minima = tabMeteo1[i][1] # atualiza se encontrar valor menor
            i = i + 1
    return minima

def amplTerm(tabMeteo1): 
    res = []
    for data, tmin, tmax, prec in tabMeteo1:
        amplitude = tmax - tmin            # calcula a amplitude
        res.append((data, amplitude))      # guarda o par (data, amplitude)
    return res


def maxChuva(tabMeteo1):
    max_prec = tabMeteo1[0][3]
    max_data = tabMeteo1[0][0]
    for data, tmin, tmax, prec in tabMeteo1:
        if prec > max_prec:
            max_prec = prec
            max_data = data
    return (max_data, max_prec)


def diasChuvosos(tabMeteo, p):
    print('data                 |         precipitação')
    while i < len(tabMeteo):
        if tabMeteo[i][3] > p:
            print(f'{tabMeteo[i][0]}         |         {tabMeteo[i][3]}')
        i = i + 1
    return

def maxPeriodoCalor(tabMeteo, p):
    max_consecutivos = 0
    consecutivos_atual = 0  # Sequência atual de dias secos na tabela de registos

    for previsão in tabMeteo:
        prec = previsão[3]
        if prec < p:
            consecutivos_atual = consecutivos_atual + 1
        else:
            if consecutivos_atual > max_consecutivos:
                max_consecutivos = consecutivos_atual
            consecutivos_atual = 0  # Reiniciamos a contagem para a próxima sequência

    # Voltamos a confirmar se a última sequência foi a maior
    if consecutivos_atual > max_consecutivos:
        max_consecutivos = consecutivos_atual

    return max_consecutivos

from matplotlib import pyplot as plt

def grafTabMeteo(t):
    x = [f"{data[0]}/{data[1]}/{data[2]}" for data,tmin,tmax,prec in t ]
    y_min = [tmin for data, tmin, tmax, prec in t]
    y_max = [tmax for data, tmin, tmax, prec in t]
    prec= [prec for *_, prec in t ] # *_ representa os outros #preguiça
    
    plt.plot(x, y_min,label = "temperatura minima(ºC)" ,color = "blue", marker = "o")
    plt.plot(x, y_max, label = " temperatura maxima (ºC)", color = "red" , marker = "o")
    plt.legend()
    plt.grid()
    plt.xticks(rotation= 45)
    plt.show()
    
    
    plt.bar(x, y_min,label = "pluviosidade" ,color = "pink")
    plt.legend()
    plt.xticks(rotation= 45)
    plt.show()
    
    return


# Variável onde vamos armazenar os dados metereológicos
tab_meteo1 = [((2022,1,20), 2, 16, 0),((2022,1,21), 1, 13, 0.2), ((2022,1,22), 7, 17, 0.01)]

opção = -1

while opção != 0:
    menu()
    opção = int(input('Selecione a opção desejada: '))

    if opção == 1:
        print(medias(tab_meteo1))
    elif opção == 2:
        print(guardaTabMeteo(tab_meteo1, 'meteorologia.txt'))
    elif opção == 3:
        print(carregaTabMeteo('meteorologia.txt'))
    elif opção == 4:
        print(minMin(tab_meteo1))
    elif opção == 5:
        print(amplTerm(tab_meteo1))
    elif opção == 6:
        print(maxChuva(tab_meteo1))
    elif opção == 7:
        print(diasChuvosos(tab_meteo1, 0))
    elif opção == 8:
        print(maxPeriodoCalor(tab_meteo1, 0.6))
    elif opção == 9:
        print(grafTabMeteo(tab_meteo1))
print('Até à próxima!')
```
