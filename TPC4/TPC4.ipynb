3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);

````python
#Cinema = [sala]
#sala = [nlugares, Vendidos, filme]
#nlugares = Int
#filme = String 
#Vendidos = [Int]

#sala1 = (150, [], "Twilight")
#sala2 = (200, [], "Hannibal")
#cinema1 = []

#cinema1 = inserirSala(cinema1,sala1)
#cinema1 = inserirSala(cinema1,sala2)

#listar(cinema1)


#if(disponivel(cinema1, "Twilight", 17 )):
 # cinema1 = vendebilhete(cinema1, "Twilight", 17 )

#listardisponibilidades(cinema1)

# Criar o menu da aplicação 

def menu():
    print('\n Menu')
    print('1) Lista de filmes em exibição')
    print('2) Disponibilidade de um lugar específico na sala') # Tem de dar um boleano como resultado da função
    print('3) Bilhete vendido')
    print('4) Lugares disponíveis em cada sala')
    print('5) Procurar filme no cinema')
    print('6) Inserir uma nova sala')
    print('0) Sair')


# Funções da aplicação

#1. `listar( cinema )` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado
# como argumento;
def listar(cinema): # 'cinema' deve ser um iterável (ex.: lista/tuplo) onde cada elemento é uma 'sala'
    for sala in cinema:
        # 'sala' deve ser uma sequência (lista/tuplo) com pelo menos 3 elementos
        # o título do filme está no índice 2 (3.º elemento)
        print(sala[2])# imprime no ecrã o elemento índice 2 da sala
    return # retorna None (linha opcional — pode ser omitida)




#2. `disponivel( cinema, filme, lugar )` - que dá como resultado **False** se o lugar lugar já estiver ocupado 
# na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;

def Lugar_disponibilidade(cinema, filme, lugar):
        # Esta função verifica se um 'lugar' está livre ou ocupado
    # no filme indicado, dentro da lista 'cinema'.
    for sala in cinema: # Percorre todas as salas do cinema (cada sala é uma lista/tuplo)
        if sala[2] == filme:
            # Verifica se o filme nesta sala é o mesmo que o procurado
            # sala[2] → nome do filme
            # Se for o filme certo, vai verificar os lugares
            if lugar in sala[1]:
                # sala[1] → lista de lugares ocupados
                # Verifica se o número 'lugar' está dentro da lista de ocupados
                print(f'Lugar {lugar} está ocupado.')
                return False  # retorna False, pois o lugar já está ocupado

            elif lugar > sala[0]:
                # sala[0] → número total de lugares na sala
                # Se o número do lugar for maior que o total de lugares, é inválido
                print(f'Lugar {lugar} é inválido.')
                return False # retorna False, pois o lugar não existe

            else:
                # Se não está ocupado e é um número válido, está disponível
                print(f'Lugar {lugar} está disponível.')
                return True # retorna True, pois o lugar está livre
    # Se o ciclo acabar sem encontrar o filme procurado:
    print(f'Filme "{filme}" não disponível.')
    return False  # retorna False, pois o filme não está em exibição




#3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar
# à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
def Bilhete_vendido(cinema, filme, lugar):
     # Esta função tenta vender um bilhete (lugar) para o filme indicado.
    # Se o lugar estiver livre e for válido, adiciona-o à lista de lugares ocupados.
    # Se o lugar estiver ocupado, for inválido ou o filme não existir, mostra uma mensagem de erro.
    
    for sala in cinema: # Percorre todas as salas do cinema (cada sala é uma lista ou tuplo)
        if sala[2] == filme:
            # Verifica se o filme nesta sala é o mesmo que o filme pedido.
            # sala[2] → nome do filme nesta sala.
            if lugar not in sala[1] and lugar <= sala[0]:
                # Verifica duas condições AO MESMO TEMPO:
                # 1️'lugar not in sala[1]' → o lugar ainda não está ocupado.
                # 2️ 'lugar <= sala[0]' → o número do lugar é válido (não ultrapassa o total de lugares).
                sala[1].append(lugar)
                # Adiciona o número do lugar à lista de lugares ocupados dessa sala.
                # Isto modifica diretamente a lista original 'cinema'.
                print(f'Lugar {lugar} vendido com sucesso para o filme "{filme}"!')
                return  # Sai imediatamente da função (não precisa continuar a procurar).
            else:
                # Este 'else' executa se:
                # - o lugar já estiver ocupado (ou seja, está em sala[1])
                # - ou o lugar for maior que o total de lugares (inválido)
                print(f'Lugar {lugar} já está ocupado ou é inválido.')
                return
                # Sai da função, pois já encontrou o filme e tratou o caso.

    # Se o ciclo 'for' terminar sem encontrar o filme (nenhum sala[2] == filme):
    print(f'Filme "{filme}" não disponível')
    # Informa o utilizador que o filme não está em exibição.
    
    
    
#Resumo do funcionamento passo a passo
#Percorre todas as salas do cinema.
#Quando encontra a sala onde o filme está a ser exibido:
#Verifica se o lugar está livre e é válido.
#Se sim → adiciona o lugar à lista de ocupados e mostra mensagem de sucesso.
#Se não → avisa que o lugar está ocupado ou inválido.
#Se não encontrar o filme em nenhuma sala, mostra que o filme não está disponível.



#4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que
# está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número
# de lugares ocupados);

# Opção 4 --> Listar lugares disponíveis em cada sala
def Listar_disponibilidade(cinema):
     # Esta função mostra no ecrã o número de lugares disponíveis
    # em cada sala do cinema.
    print('Lugares disponíveis')
     # Cabeçalho informativo — apenas texto para o utilizador perceber o que está a ver.
    print('Filmes ------- Lugares')
    # Outra linha de cabeçalho que mostra como os dados serão apresentados:
    # nome do filme seguido do número de lugares disponíveis.

    for sala in cinema:
        # Percorre todas as salas na lista 'cinema'.
        # Cada 'sala' é uma lista (ou tuplo) com:
        # sala[0] → total de lugares
        # sala[1] → lista de lugares ocupados
        # sala[2] → nome do filme
        lug_disponíveis = sala[0] - len(sala[1])
        # Calcula o número de lugares disponíveis:
        # total - ocupados = disponíveis
        # Exemplo: 10 - 3 = 7 lugares livres
        print(f'{sala[2]} ------- {lug_disponíveis}')
               # Mostra o nome do filme (sala[2]) e o número de lugares disponíveis.
        # O '-------' é apenas para separar visualmente no ecrã.

    return
    # A função termina aqui. O 'return' é opcional porque não devolve nada (apenas imprime).



#5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);


# Opção 5 --> Procurar um filme em exibição no cinema
def Procura(cinema, filme):
     # Função que procura se um determinado filme está em exibição no cinema.
    # 'cinema' é uma lista de salas, e 'filme' é o nome do filme a procurar.
    
    res = -1
    # Guarda o resultado da pesquisa. Se o filme não for encontrado, devolve -1.
    i = 0
        # Inicializa a variável 'i' (o índice da sala) em 0.
    for i, sala in enumerate(cinema): # i corresponde ao índice de cada sala dentro do cinema
        # O 'enumerate' percorre a lista 'cinema' e devolve, em cada iteração:
        # i → o número da sala (índice na lista)
        # sala → a lista que contém [lugares_totais, lugares_ocupados, nome_do_filme]
        if sala[2] == filme:
             # Verifica se o nome do filme na sala é igual ao filme procurado.
            print(f'Filme "{filme}" encontrado na sala {i + 1}.') 
            # Mostra a mensagem com o número da sala (i + 1, porque o índice começa em 0).
            return i
            # Devolve o índice da sala onde o filme foi encontrado.
    
    print(f'Filme "{filme}" não encontrado.') # Se o ciclo terminar sem encontrar o filme, mostra esta mensagem.
    return res
    # Retorna -1, indicando que o filme não foi encontrado.



#6. Acrescente todas as outras funcionalidades que achar necessárias;
# Opção 6 --> Inserir nova sala, garantindo que não haja duplicação de salas

def Inserir_sala(cinema, sala):
    
    for s in cinema:
        if s[2] == sala[2]:  # Verifica se o filme já está associado a outra sala
            print(f'A sala com o filme "{sala[2]}" já existe.')
            return cinema
    cinema.append(sala)
    print(f'Sala adicionada com sucesso: {sala}')
    return cinema

#7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.
# ---------------------------- Programa Principal ----------------------------

# Dados do cinema
cinema = [
    [200, [], 'Shrek 2'],
    [150, [], 'How To Lose a Guy In 10 Days'],
    [250, [], 'La La Land'],
    [200, [], 'Interstellar']
]

opção = -1
while opção != 0:
    menu()
    opção = int(input('Selecione a opção desejada: '))
    
    if opção == 1:
        listar(cinema)
    elif opção == 2:
        filme = input('Introduza o nome do filme: ')
        lugar = int(input('Introduza o número do lugar: '))
        Lugar_disponibilidade(cinema, filme, lugar)
    elif opção == 3:
        filme = input('Introduza o nome do filme: ')
        lugar = int(input('Introduza o número do lugar que deseja: '))
        Bilhete_vendido(cinema, filme, lugar)
    elif opção == 4:
        Listar_disponibilidade(cinema)
    elif opção == 5:
        filme = input('Introduza o nome do filme: ')
        Procura(cinema, filme)
    elif opção == 6:
        num_lugares = int(input('Número de lugares da nova sala: '))
        filme = input('Nome do novo filme a exibir: ')
        nova_sala = [num_lugares, [], filme]
        cinema = Inserir_sala(cinema, nova_sala)
print('Até à próxima!')
````
